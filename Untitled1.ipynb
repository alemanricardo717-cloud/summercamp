```python
import pandas as pd
import sqlite3
import seaborn as sns
import matplotlib.pyplot as plt

# 1. Creamos datos de ejemplo nuevos
datos_tienda = {
    'Producto': ['PS5', 'Xbox Series X', 'Nintendo Switch', 'PC Gamer', 'Steam Deck'],
    'Precio_USD': [500, 480, 300, 1200, 400],
    'Cantidad_Stock': [15, 10, 25, 5, 8]
}

df_tienda = pd.DataFrame(datos_tienda)

# 2. Conexión a SQL y guardar tabla
conn = sqlite3.connect("tienda.db")
df_tienda.to_sql("inventario", conn, if_exists="replace", index=False)

# 3. Consulta SQL para calcular el Valor Total (Precio * Cantidad)
query = """
SELECT Producto, (Precio_USD * Cantidad_Stock) AS Valor_Total
FROM inventario
ORDER BY Valor_Total DESC
"""
df_resultado = pd.read_sql_query(query, conn)

# 4. Graficar los resultados
plt.figure(figsize=(10, 5))
sns.barplot(data=df_resultado, x="Producto", y="Valor_Total", palette="viridis", hue="Producto", legend=False)

plt.title("💰 Valor Total del Inventario por Consola", fontsize=14)
plt.ylabel("Valor en USD")
plt.show()
```
